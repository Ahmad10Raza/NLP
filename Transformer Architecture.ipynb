# **Transformer Architecture**

# What is Transformer Architecture?

The Transformer architecture is a deep learning model introduced in the paper "Attention Is All You Need" by Vaswani et al. in 2017. It has revolutionized the field of natural language processing (NLP) and become the foundation for many state-of-the-art NLP models, such as BERT, GPT, and more. The Transformer architecture is not limited to NLP and can be applied to various sequence-to-sequence tasks, including machine translation, text summarization, and more.

Key features of the Transformer architecture:

1. **Self-Attention Mechanism**:
   - The core innovation in the Transformer is the self-attention mechanism. It allows the model to weigh the importance of different words in a sentence, considering their context. This mechanism is crucial for capturing long-range dependencies in sequences.

2. **Parallelization**:
   - Unlike recurrent neural networks (RNNs) and convolutional neural networks (CNNs), the Transformer architecture can process input sequences in parallel. This greatly speeds up training and inference.

3. **Encoder-Decoder Architecture**:
   - The Transformer is typically used in a sequence-to-sequence setting, where an encoder processes the input sequence and a decoder generates the output sequence.

4. **Positional Encoding**:
   - Since the Transformer doesn't have an inherent sense of the order of elements in a sequence (unlike RNNs), positional encoding is added to the input to provide information about the position of words in the sequence.

5. **Multi-Head Attention**:
   - To capture different types of relationships in the data, the Transformer uses multi-head attention. It performs multiple attention operations in parallel, each focusing on different aspects of the input.

6. **Feedforward Neural Networks**:
   - After the attention mechanism, the Transformer includes feedforward neural networks to further process the information.

7. **Layer Normalization**:
   - Layer normalization is applied after each sub-layer within the model to stabilize training.

8. **Stacking Layers**:
   - Transformers are typically deep models, consisting of multiple layers of encoders and decoders. Stacking layers allows the model to learn hierarchical representations.

9. **Masking**:
   - In tasks like machine translation, masking is used to prevent the decoder from attending to future tokens, ensuring causal attention and autoregressive behavior.

The Transformer architecture has several advantages, including its ability to capture long-range dependencies, its parallelism, and its scalability to handle large datasets. This architecture has become the foundation for a wide range of NLP models, leading to significant improvements in tasks like translation, summarization, question-answering, and sentiment analysis.

Since its introduction, the Transformer has been adapted and extended into various variants, such as the BERT model for pre-training and GPT (Generative Pre-trained Transformer) for generative tasks, further demonstrating its versatility and effectiveness in natural language understanding and generation tasks.

The Transformer architecture relies on a self-attention mechanism and consists of several key components. Here's an explanation of its architecture and components:

**Components of the Transformer Architecture**:

1. **Input Embeddings**:
   - The input to the Transformer consists of a sequence of tokens (e.g., words or subwords). Each token is initially represented as an embedding vector, which can be learned during training or obtained from pre-trained word embeddings.

2. **Positional Encoding**:
   - Since the Transformer does not have an inherent sense of word order like recurrent neural networks (RNNs), positional encodings are added to the input embeddings. These encodings provide information about the position of tokens in the sequence.

3. **Multi-Head Self-Attention Mechanism**:
   - The self-attention mechanism is at the core of the Transformer. It allows the model to weigh the importance of different tokens in the input sequence when making predictions.
   - Multi-head attention involves multiple attention mechanisms running in parallel, each focusing on different aspects of the input.
   - The self-attention mechanism computes attention scores, which are used to determine how much focus should be given to each token in the sequence.

4. **Layer Normalization**:
   - Layer normalization is applied after the self-attention mechanism and the feedforward neural networks. It helps stabilize training and improves the flow of gradients through the model.

5. **Feedforward Neural Networks (FFN)**:
   - After self-attention, the model passes the outputs through a feedforward neural network for further processing and feature extraction.
   - FFNs are typically applied independently to each position in the sequence.

6. **Encoder-Decoder Architecture (For Sequence-to-Sequence Tasks)**:
   - In tasks like machine translation, the Transformer uses an encoder-decoder architecture.
   - The encoder processes the input sequence, while the decoder generates the output sequence.

7. **Output Embeddings**:
   - The final layer of the decoder produces output embeddings, which are then converted into probabilities for each token in the output vocabulary. These probabilities are used to generate the output sequence.

8. **Position-wise Feedforward Networks**:
   - The FFNs in the Transformer operate on a position-wise basis, meaning they process each token's representation independently.

9. **Residual Connections**:
   - To facilitate the training of very deep networks, residual connections are employed. These connections allow gradients to flow more effectively during backpropagation.

10. **Stacked Layers**:
   - Transformers are typically deep models with multiple layers of encoders and decoders, allowing them to learn hierarchical representations of the input data.

The general architecture of the Transformer includes multiple layers of encoders and decoders, each consisting of the components mentioned above. The self-attention mechanism plays a crucial role in modeling relationships and dependencies between tokens in the sequence, making it well-suited for a wide range of NLP tasks.

Variants and extensions of the Transformer architecture have been developed for specific tasks, such as the BERT model for pre-training and GPT (Generative Pre-trained Transformer) for generative tasks. These models have achieved remarkable performance across various NLP applications.

# Notation Used In `Attention Is All You Need`

These notations include:

   - **Value (V)**: The value vector is represented as `V`.
   - **Attention Scores (A)**: The attention scores are often represented as `A`.
   - **Attention Weights (W)**: The attention weights are often represented as `W`.

4. **Multi-Head Attention**:
   - **Number of Heads**: The number of attention heads is typically represented as `h`.
   - **Attention for Head `i`**: To indicate the attention scores or weights for a specific head, a subscript is often used, such as `A_i` or `W_i`.

5. **Layer Notation**:
   - **Layer Index**: The index for a specific layer is often represented as `l`.

6. **Feedforward Neural Networks (FFN)**:
   - **Output of FFN**: The output of the FFN is often represented as `FFN(x)`.

7. **Residual Connections and Layer Normalization**:
   - **Residual Connection**: Residual connections are denoted by adding the input to the output, such as `x + FFN(x)`.
   - **Layer Normalization**: Layer normalization can be indicated with the notation `LN(x)`.

8. **Position-wise Processing**:
   - To represent the processing of each token independently in position-wise processing, subscripts are often used, such as `x_i` or `FFN(x)_i`.

9. **Stacked Layers**:
   - To represent stacked layers in the model, layer indices are typically used, such as `l_1`, `l_2`, etc.

10. **Encoder and Decoder**:
    - **Encoder Output**: The output of the encoder is often represented as `Encoder(X)`.
    - **Decoder Output**: The output of the decoder is often represented as `Decoder(Y)`.

11. **Subword Segmentation**:
    - To represent subword segmentation, notations like `` (underscore) are used to indicate spaces in subword tokens, e.g., "undoable."

These are the standard notations used in the original Transformer architecture and many subsequent models based on it. These notations help in describing the architecture and components of the model more precisely and are widely adopted in the literature.

## What is an input embedding?

Input embedding is a fundamental concept in natural language processing and deep learning. It involves representing words or tokens in a numerical format that is suitable for processing by neural networks. To explain input embedding mathematically, let's break it down step by step.

````python
img3=Image.open('Data/Images/Input_Embeddings.png')
img3=img3.convert('RGB')
display(img3)
````

**Input Embedding**:

1. **Word Representation**:
   - In natural language processing, text data is often represented as sequences of words or tokens. Each word needs to be converted into a numerical form so that it can be used as input to a neural network.

2. **One-Hot Encoding (Conceptual Explanation)**:
   - One simple way to represent words numerically is through one-hot encoding. In this approach, each word in the vocabulary is associated with a unique integer index. For a given word, its one-hot encoding is a binary vector where all elements are zero except for the one corresponding to the word's index. Mathematically, this can be represented as follows:

     - Suppose we have a vocabulary V with N unique words. The one-hot encoding of a word w_i would be a vector of length N, with a 1 at index i and 0s elsewhere:
     
       ```
       w_i = [0, 0, ..., 1, ..., 0]
                  
                  i-th position
       ```

3. **Word Embeddings (Continuous Representation)**:
   - One-hot encoding has limitations in capturing semantic relationships between words. Word embeddings offer a more expressive representation. Word embeddings are continuous vector representations of words in a continuous vector space.
   - Word embeddings are typically learned during training, and the embedding vectors capture semantic similarities between words. Mathematically, this can be represented as follows:

     - Suppose we have a word embedding matrix E, where each row corresponds to the embedding vector of a word. For a given word w_i, its word embedding can be represented as a vector e_i:

       ```
       e_i = E[i]
       ```

   - The word embeddings are learned in a way that similar words have similar embeddings, allowing neural networks to capture semantic relationships between words.

4. **Sentence or Sequence Embedding**:
   - In the context of NLP, sentences or sequences are typically represented as a series of word embeddings. These word embeddings are often stacked together to form a sequence or sentence embedding.

   - For a sequence of words, denoted as [w_1, w_2, ..., w_n], the corresponding sequence embedding can be represented as follows:

     ```
     Sentence Embedding = [e_1, e_2, ..., e_n]
     ```

   - The sequence embedding is then used as input to a neural network for further processing, such as in the Transformer architecture.

Input embedding, as described, allows neural networks to work with text data efficiently and capture the semantic information of words within a continuous vector space, facilitating various natural language processing tasks.

## What is positional encoding?

**Positional encoding is a crucial component in the Transformer architecture used for natural language processing tasks. Its primary purpose is to provide information about the position of tokens in a sequence, addressing the challenge that Transformers don't have an inherent understanding of token order. Here's an explanation of positional encoding, along with the mathematical concept and its usage in the Transformer architecture:**

**Concept of Positional Encoding**:

In a sequence of tokens, tokens are order-dependent, and their positions in the sequence carry essential information. Traditional models like RNNs and LSTMs inherently capture sequential information, but Transformers process tokens in parallel and do not have this built-in understanding of sequence order. Positional encoding is introduced to rectify this limitation.

**Mathematical Concept of Positional Encoding**:

1. A common approach to positional encoding is to use trigonometric functions like sine and cosine to create position-specific embeddings for tokens.
2. Positional encodings are usually additive, meaning they are added to the word embeddings to capture the position-based information.
3. For each position in the sequence, positional encodings are calculated based on the position and the dimension of the embedding space.

The mathematical representation of positional encoding for a given position `pos` and dimension `i` in a sequence is as follows:

```python
PE(pos, 2i) = sin(pos / 10000^(2i / d_model))
PE(pos, 2i+1) = cos(pos / 10000^(2i / d_model))
```

Where:
- `pos` is the position of the token in the sequence.
- `i` is the dimension index.
- `d_model` is the dimension of the word embeddings (e.g., 512 in the original Transformer).

The idea behind this encoding is that the sine and cosine functions with different frequencies allow the model to capture different positional information in different dimensions, making the positions distinguishable in the embedding space.

**Usage in Transformer Architecture**:

In the Transformer architecture, positional encoding is added to the word embeddings before feeding them into the model. This is done in both the encoder and decoder components.

The positional encoding vectors are constructed for each position in the sequence and each dimension of the embedding space, according to the mathematical concept described above. These positional encodings are added to the corresponding word embeddings.

The reason positional encoding is used in the Transformer is to allow the model to distinguish between tokens based on their position in the sequence. Without positional encoding, the model would not have any inherent information about the order of the tokens. By adding positional encoding, the Transformer can effectively process sequences and capture sequential relationships, making it suitable for tasks like language understanding, machine translation, and text generation.

## Why trigonometric functions?

Trigonometric functions, specifically the sine and cosine functions, are used in positional encoding in the Transformer architecture due to several important properties that make them well-suited for this purpose. The choice of sine and cosine functions is not arbitrary and is based on their desirable characteristics. Here's why sine and cosine functions are used instead of other trigonometric functions:

1. **Periodicity**:
   - Sine and cosine functions are periodic, which means they repeat their values at regular intervals. This property is essential for positional encoding because it allows the model to capture the positional information for tokens at different positions in the sequence.
   - The periodicity of sine and cosine functions ensures that tokens at different positions receive unique positional embeddings.

2. **Orthogonality**:
   - Sine and cosine functions are orthogonal, meaning they are uncorrelated with each other. This property is beneficial for creating embeddings that capture different aspects of positional information.
   - The combination of sine and cosine functions at different frequencies provides a rich set of positional encodings, allowing the model to distinguish between positions effectively.

3. **Amplitude and Phase Shift**:
   - Sine and cosine functions have different amplitudes and phase shifts, which are crucial for creating diverse positional embeddings. This diversity is essential for the model to differentiate between positions.
   - By using sine and cosine functions with different frequencies and phase shifts, the model can encode information about the order of tokens in the sequence without introducing collinearity.

4. **Easy to Compute**:
   - Sine and cosine functions are computationally efficient and easy to implement. They can be efficiently computed in parallel, making them practical for deep learning models like the Transformer.

5. **Interpretable**:
   - Sine and cosine functions are easily interpretable. The model can learn to assign specific frequencies to different dimensions, which can help understand the encoding scheme better and provide insight into how positional information is captured.

While sine and cosine functions are the most commonly used functions for positional encoding in the Transformer, other functions can theoretically be used for this purpose. However, the unique combination of periodicity, orthogonality, diversity, and computational efficiency makes sine and cosine functions a compelling choice for encoding positional information in a sequence of tokens. These functions allow the Transformer model to effectively capture and differentiate positional information while maintaining efficiency and interpretability.

# What is Self-Attention & How to compute Self-Attention ?

Self-attention is a mechanism that allows a neural network to weigh the importance of different elements in a sequence when making predictions or extracting features. It is a key component of the Transformer architecture and has been instrumental in improving the performance of natural language processing tasks. Self-attention is often used to capture dependencies between elements in a sequence, making it suitable for tasks like language modeling, machine translation, and text generation.

Here's how to compute self-attention:

**1. Key, Query, and Value**:

In self-attention, you have three sets of vectors: keys, queries, and values, which are derived from the input sequence:

- **Key (K)**: Key vectors represent elements in the input sequence and are used to determine how much focus should be given to other elements.

- **Query (Q)**: Query vectors represent the elements for which you want to compute attention scores, indicating how relevant each element is to the query.

- **Value (V)**: Value vectors contain the actual information associated with each element.

These key, query, and value vectors are computed through linear transformations of the input embeddings (or hidden states) and are used to compute attention scores.

**2. Compute Attention Scores**:

The attention score between a query (Q) and a key (K) is typically calculated using the dot product or another similarity metric:

- **Dot Product (Scaled)**: The scaled dot-product attention is a common choice. The attention score is computed as follows:

  ```
  Attention(Q, K) = softmax((Q * K^T) / sqrt(d_k))
  ```

  Here, `*` represents the dot product, `^T` denotes the transpose, and `d_k` is the dimension of the key vectors. The division by `sqrt(d_k)` is used to scale the dot products, which helps stabilize training.

**3. Normalize Attention Scores**:

Apply the softmax function to the attention scores to get attention weights, ensuring they sum to 1:

- `Attention Weights = softmax(Attention Scores)`

**4. Calculate Weighted Sum**:

Compute the weighted sum of the value vectors using the attention weights to obtain the context vector. The context vector represents the information from the input sequence that is relevant to the query:

- `Context Vector = sum(Attention Weights * Values)`

The computed context vector contains information from the input sequence, with different weights assigned to each element based on their relevance to the query.

**5. Multiple Heads (Optional)**:

In practice, you may use multiple sets of key, query, and value vectors, each called an "attention head." This allows the model to learn different relationships and dependencies in parallel.

The self-attention mechanism is used to capture complex patterns and relationships in sequential data, and it's a key building block of models like the Transformer. By computing attention scores between different elements in the input sequence, the model can learn to focus on the most relevant parts of the sequence for various tasks, such as translation, summarization, and text understanding.

# Multi-head Attention

Multi-head attention is a mechanism used in deep learning models, particularly in the Transformer architecture, to capture complex relationships and dependencies in sequential data. It involves running multiple sets of self-attention mechanisms in parallel, each focusing on different aspects of the input sequence. Here, I'll explain multi-head attention with a mathematical concept.

**Mathematical Concept of Multi-Head Attention**:

In multi-head attention, you compute attention scores, attention weights, and context vectors for each attention head separately, using different learned parameters for each head. The outputs from each head are then combined to produce the final multi-head attention output.

Here's a step-by-step mathematical concept of multi-head attention:

1. **Initialization**:
   - Let's assume we have `H` attention heads. For each head `i`, you calculate separate sets of key (K_i), query (Q_i), and value (V_i) vectors. These projections are obtained by linear transformations of the input sequence.

2. **Attention Score Calculation for Each Head**:
   - For each attention head `i` (where `i` ranges from 1 to `H`), calculate attention scores (Attention_i) independently. The attention scores represent the relevance of each element in the sequence to the query vectors for this head.

   ```
   Attention_i(Q_i, K_i) = softmax((Q_i * K_i^T) / sqrt(d_k))
   ```

   Here:
   - `Q_i` is the query vector for head `i`.
   - `K_i` is the key vector for head `i`.
   - `^T` denotes the transpose.
   - `d_k` is the dimension of the key vectors, which is used to scale the dot products.

3. **Context Vector Calculation for Each Head**:
   - Use the attention scores (Attention_i) to compute context vectors (Context_i) for each attention head `i`:

   ```
   Context_i = Attention_i * V_i
   ```

   Here:
   - `V_i` is the value vector for head `i`.

4. **Combination of Context Vectors**:
   - The context vectors from each attention head represent different aspects of the input sequence and different relationships. These context vectors are often combined to obtain the multi-head attention output:

   ```
   Multi-Head Attention Output = Concatenate(Context_1, Context_2, ..., Context_H) * W_O
   ```

   Here:
   - `W_O` is a learned weight matrix applied to the concatenated context vectors to produce the final output.

The multi-head attention mechanism allows the model to capture diverse and complex patterns in the input sequence. Each attention head can focus on different relationships, and the combination of their outputs provides a rich representation of the data. This makes the model more flexible and enables it to handle various sequence processing tasks effectively.

This concept is central to the success of the Transformer architecture and is used for a variety of reasons:

**1. Capturing Diverse Relationships**:
   - The self-attention mechanism captures relationships between tokens in a sequence. By using multiple attention heads, the model can learn and capture diverse types of relationships. Each head may specialize in different relationships, such as local dependencies or long-range dependencies, making the model more flexible and expressive.

**2. Learning Multiple Representations**:
   - Each attention head produces its own set of attention scores and context vectors. These multiple context vectors can be combined to provide multiple representations of the same input sequence. This can capture different aspects of the data and enrich the model's understanding.

**3. Reducing Attention Overhead**:
   - In deep models, especially those with long sequences, a single self-attention mechanism could be computationally expensive. Multi-head attention can alleviate this by distributing the attention computation across multiple heads. This results in a more efficient model while maintaining its ability to capture dependencies effectively.

**4. Improved Generalization**:
   - Multi-head attention enables the model to generalize better. Different attention heads can focus on different parts of the input sequence. During training, the model learns which heads are more useful for specific tasks, allowing it to adapt to the unique characteristics of different inputs.

**5. Redundancy Reduction**:
   - Using multiple attention heads can reduce the risk of overfitting and redundancy in the learned representations. Each head focuses on a distinct aspect, reducing the chance of all heads learning the same patterns.

**6. Interpretable Models**:
   - Multi-head attention can provide interpretability. Analysts can examine the attention weights from different heads to gain insights into which parts of the input sequence are most relevant for the model's predictions.
